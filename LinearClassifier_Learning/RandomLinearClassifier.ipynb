```python
import random
from Tools import linearClassifyingTools

def randomGen(d):
    theta = [random.randint(-10,10) for i in range(d)]
    if 0 in theta:
        theta = [random.randint(-10,10) for i in range(d)]

    theta_zero = random.randint(-10,10)
    
    return theta,theta_zero
```

```python
def linearClassifier(D,k,d):
    parameters= []
    losses = []

    # print("started")
    
    for j in range(k):
        theta, theta_zero = randomGen(d)
        parameters.append((theta, theta_zero))

        #Find errors
        loss = linearClassifyingTools.findLoss(theta, theta_zero,D)
        losses.append(loss)

        # print("loss is " + str(loss))
   

    bestIndex = losses.index(min(losses))

    # print("Best loss was " + str(losses[bestIndex]))
    
    return parameters[bestIndex][0],parameters[bestIndex][1],losses[bestIndex], losses
```
